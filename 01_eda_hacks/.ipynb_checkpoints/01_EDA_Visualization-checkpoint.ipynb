```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,'lifespan': lifespan}, index=index)
df
```

```python
ax = df.plot.bar(rot=0)
```

```python
ax = df.plot.bar(stacked=True)
```

```python
axes = df.plot.bar(rot=0, subplots=True, color={"speed": "red", "lifespan": "green"})
axes[1].legend(loc=2);
```

# KDE
In statistics, [kernel density estimation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.kde.html) (KDE) is a non-parametric way to estimate the probability density function (PDF) of a random variable <br>
This function uses Gaussian kernels and includes automatic bandwidth determination.

```python
s = pd.Series([1, 2, 2.5, 3, 3.5, 4, 5])
ax = s.plot.kde()
```

```python
df = pd.DataFrame({
    'x': [1, 2, 2.5, 3, 3.5, 4, 5],
    'y': [4, 4, 4.5, 5, 5.5, 6, 6],
})

ax = df.plot.kde()
```

```python
ax = df.plot.kde(bw_method=0.3)
```

# Hexabin

```python
n = 500

df = pd.DataFrame({
    'coord_x': np.random.uniform(-3, 3, size=n),
    'coord_y': np.random.uniform(30, 50, size=n),
    'observations': np.random.randint(1,5, size=n)
    })

ax = df.plot.hexbin(x='coord_x',
                    y='coord_y',
                    C='observations',
                    reduce_C_function=np.sum,
                    gridsize=10,
                    cmap="viridis")
```
